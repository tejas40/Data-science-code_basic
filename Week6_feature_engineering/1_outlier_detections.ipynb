# how to detect outlier using quantile

```python
df=pd.read_csv("heights.csv")
```

```python
df
```

```python
max_thresh=df["height"].quantile(0.90)
```

```python
df[df["height"]>max_thresh]
```

```python
min_thresh=df["height"].quantile(0.05)
```

```python
df[df["height"]<min_thresh]
```
