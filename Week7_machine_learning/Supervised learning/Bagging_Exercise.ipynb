```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/19_Bagging/Exercise/heart.csv")
```

```python
df1 = df[df["RestingBP"]<(df["RestingBP"].mean()+3*df["RestingBP"].std())]
df2 = df1[df1["Cholesterol"]<(df1["Cholesterol"].mean()+3*df1["Cholesterol"].std())]
df3 = df2[df2["Oldpeak"]<(df2["Oldpeak"].mean()+3*df2["Oldpeak"].std())]
df4 = df3.copy()
df4
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
df4["Sex"] =encode.fit_transform(df4["Sex"])
df4["ExerciseAngina"] =encode.fit_transform(df4["ExerciseAngina"])
df4
```

```python
df4["ChestPainType"].unique()
```

```python
df4["RestingECG"].unique()
```

```python
df4["ST_Slope"].unique()
```

```python
df4 = pd.get_dummies(df4 , columns=["ChestPainType","RestingECG","ST_Slope"], drop_first=True)
```

```python
df4
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
df4[["Age","RestingBP","Cholesterol","MaxHR","Oldpeak"]]= scale.fit_transform(df4[["Age","RestingBP","Cholesterol","MaxHR","Oldpeak"]])
```

```python
df4
```

```python
df4.describe()
```

```python
X = df4.drop("HeartDisease",axis=1)
y= df4["HeartDisease"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
model.score(X_test, y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
model1= BaggingClassifier(estimator=SVC() ,n_estimators=100 ,max_samples=0.5, max_features=0.5, oob_score=True, random_state=0)
model1.fit(X_train, y_train)
model1.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier()
model2.fit(X_train, y_train)
model2.score(X_test, y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
model3= BaggingClassifier(estimator=DecisionTreeClassifier() ,n_estimators=100 ,max_samples=0.3, max_features=0.3, oob_score=True, random_state=0)
model3.fit(X_train, y_train)
model3.score(X_test,y_test)
```

```python
model1.oob_score_ , model3.oob_score_
```
