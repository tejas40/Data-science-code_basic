```python
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay

usg=CustomBusinessDay(calendar=USFederalHolidayCalendar())
```

```python
usg
```

```python
rng=pd.date_range(start="8/1/2016",end="8/31/2016", freq=usg )
```
