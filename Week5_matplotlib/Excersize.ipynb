```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[32,44,50,13]
x_axis=np.arange(len(company))
```

```python
plt.bar(x_axis - 0.2 , revenue , width=0.4, label="Revenue")
plt.bar(x_axis + 0.2 , profit , width=0.4, label="profit")
plt.ylabel("Revenue(Bin)")
plt.title("US Technology Stocks")
plt.legend()
plt.savefig("Us.png", bbox_inches="tight" , pad_inches=4 )
plt.show()
```
