```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
x_axis=np.arange(len(company))
```

```python
plt.bar(x_axis, revenue)
plt.xticks(x_axis , company)
plt.xlabel("Company names")
plt.ylabel("Revenue")
plt.title("Renues of our company")
plt.savefig("bar.pdf" , bbox_inches="tight", pad_inches =1 , transparent =True)
plt.show()
```
