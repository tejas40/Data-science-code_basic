```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
plt.legend(["max"," min" , "avg"], ncol=3 , facecolor="#fff453", loc="upper right",
           shadow=True,fontsize="small")
plt.grid()
plt.show()
```
