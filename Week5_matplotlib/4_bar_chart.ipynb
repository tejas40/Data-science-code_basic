```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=["GOOGLE","AMZN","MSFT","FB"]
revenue=[90,136,89,27]
profit=[40,2,34,12]
x_axis=np.arange(len(company))
x_axis
```

```python
plt.bar(x_axis-0.2, revenue ,label="Revenue" , width=0.4)
plt.bar(x_axis+0.2, profit , label="profit", width=0.4)
plt.xticks(x_axis, company)
plt.xlabel("comapnies")
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.legend()
```

```python
plt.barh(x_axis-0.2, revenue ,label="Revenue" , height=0.4)
plt.barh(x_axis+0.2 , profit , label="profit", height=0.4)
plt.yticks(x_axis, company)
plt.xlabel("revenu")
plt.ylabel("companies")
plt.title("US Tech Stocks")
plt.legend()
```
